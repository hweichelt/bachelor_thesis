```python
from pathlib import Path

import matplotlib.pyplot as plt
import pandas as pd
import os
```

# Single MUC Benchmarks

## Assumption Scaling

```python
stats_path = Path("statistics_assumption_scaling")

df = pd.DataFrame()
for file in os.listdir(stats_path):
    if file.endswith(".csv"):
        file_df = pd.read_csv(stats_path.joinpath(file))
        df = pd.concat([df, file_df])
df
```

```python
id_single_df = df[df["algorithm"] == "id_single"].sort_values(by=['benchmark assumptions'])
bf_single_df = df[df["algorithm"] == "bf_single"].sort_values(by=['benchmark assumptions'])
bfi_single_df = df[df["algorithm"] == "bfi_single"].sort_values(by=['benchmark assumptions'])
# id_single_df = id_single_df[id_single_df["benchmark assumptions"] < 100]
id_single_df
```

```python
plt.plot(id_single_df["benchmark assumptions"], id_single_df["computation time"])
plt.plot(bf_single_df["benchmark assumptions"], bf_single_df["computation time"])
plt.plot(bfi_single_df["benchmark assumptions"], bfi_single_df["computation time"])
```

```python
plt.plot(id_single_df["benchmark assumptions"], id_single_df["computation time"])
```
